```python
from requests_html import HTMLSession
import re

url = 'https://www.ptt.cc/bbs/hotboards.html'

# 建立 HTML 會話
session = HTMLSession()
response = session.get(url)

# 檢查是否成功取得網頁
if response.status_code == 200:
    # 使用 CSS 選擇器來提取標題、推文數
    items = response.html.find('.b-ent')

    # 計數器
    count = 0

    for item in items:
        title = item.find('.board-title', first=True).text
        nuser = item.find('.board-nuser .hl', first=True).text

        # 使用正則表達式清洗標題和推文數
        cleaned_title = re.sub(r'\[.*?\]', '', title)  # 移除中括號及內容
        cleaned_nuser = re.sub(r'\s+', '', nuser)  # 移除所有空白

        # 打印清理後的標題和推文數
        print("標題:", cleaned_title.strip())
        print("推文數:", cleaned_nuser)
        print("-----")

        # 每打印一次就將計數器加一
        count += 1

        # 如果計數器達到10，就跳出迴圈
        if count == 10:
            break

else:
    print("無法取得網頁內容")

```

```python
from requests_html import HTMLSession
import re
import pandas as pd

url = 'https://www.ptt.cc/bbs/hotboards.html'

# 建立 HTML 會話
session = HTMLSession()
response = session.get(url)

# 檢查是否成功取得網頁
if response.status_code == 200:
    # 使用 CSS 選擇器來提取標題、推文數和日期
    items = response.html.find('.b-ent')

    # 儲存清理後的資料的列表
    data = []

    # 計數器
    count = 0

    for item in items:
        title = item.find('.board-title', first=True).text
        nuser = item.find('.board-nuser .hl', first=True).text

        # 使用正則表達式清洗標題和推文數
        cleaned_title = re.sub(r'\[.*?\]', '', title).strip()  # 移除中括號及內容
        cleaned_nuser = re.sub(r'\s+', '', nuser)  # 移除所有空白

        # 將清理後的資料加入列表
        data.append({
            '標題': cleaned_title,
            '推文數': cleaned_nuser
        })

        # 每處理一次就將計數器加一
        count += 1

        # 如果計數器達到10，就跳出迴圈
        if count == 10:
            break

    # 將資料列表轉換成 DataFrame
    df = pd.DataFrame(data)

    # 將 DataFrame 儲存為 CSV 檔案
    df.to_csv('output.csv', index=False)

    print("已儲存為 output.csv 檔案")

else:
    print("無法取得網頁內容")
```

```python
from requests_html import HTMLSession
import re
import pandas as pd
import json  # 引入 json 模組

url = 'https://www.ptt.cc/bbs/hotboards.html'

# 建立 HTML 會話
session = HTMLSession()
response = session.get(url)

# 檢查是否成功取得網頁
if response.status_code == 200:
    # 使用 CSS 選擇器來提取標題、推文數和日期
    items = response.html.find('.b-ent')

    # 儲存清理後的資料的列表
    data = []

    # 計數器
    count = 0

    for item in items:
        title = item.find('.board-title', first=True).text
        nuser = item.find('.board-nuser .hl', first=True).text

        # 使用正則表達式清潔標題和推文數
        cleaned_title = re.sub(r'\[.*?\]', '', title).strip()  # 移除中括號及內容
        cleaned_nuser = re.sub(r'\s+', '', nuser)  # 移除所有空白

        # 將清潔後的資料加入列表
        data.append({
            '標題': cleaned_title,
            '推文數': cleaned_nuser
        })

        # 每處理一次就將計數器加一
        count += 1

        # 如果計數器達到10，就跳出迴圈
        if count == 10:
            break

    # 將資料列表儲存為 JSON 檔案
    with open('output.json', 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print("已儲存為 output.json 檔案")

else:
    print("無法取得網頁內容")
```
